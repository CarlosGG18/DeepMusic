```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from deepnote import MusicRepr, Constants, Metric, Note, utils
from deepnote.metrics import (
    pitch_histogram_entropy, 
    polyphony, 
    polyphony_rate, 
    pitch_in_scale_rate, 
    empty_beat_rate,
    grooving_pattern_similarity,
    chord_progression_irregularity
)
```

```python
const = Constants(
    unit=4,
    min_tempo= 20,
    max_tempo = 200,
    num_tempo_bins=20,
    num_velocity_bins=20
)
seq = MusicRepr.from_file('test.mid', const=const)
```

```python
ent, probs = pitch_histogram_entropy(seq, window=10, pitch_class=True)

plt.figure(figsize=(20,8))
plt.subplot(2,1,1)
plt.title('entropy')
plt.plot(ent)
plt.subplot(2,1,2)
plt.title('probs')
plt.imshow(probs)
plt.show()
```

```python
polyphony(seq), polyphony_rate(seq)
```

The first call to polyphony(seq) calculates the polyphony of the music sequence represented by seq. The value returned, 3.488306451612903, indicates the average number of simultaneous notes or voices sounding at any given time in the music.

The second call to polyphony_rate(seq) calculates the polyphony rate of the music sequence represented by seq. The value returned, 0.9748427672955975, represents the ratio of the total duration of simultaneous notes to the total duration of the music sequence. It provides an indication of how densely packed the notes are in the music.
